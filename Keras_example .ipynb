```python
import numpy as py
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataset = py.loadtxt('/projects/250c93cf-311b-4af0-b0c9-a55b21af396c/data/pima-indians-diabetes.data.csv', delimiter=',')
```

```python
x = dataset[:,0:8]
y = dataset[:,8]
```

```python
x.shape[1]
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(50, input_dim=8, activation='relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x, y, epochs=8000, batch_size=7000) 
```
